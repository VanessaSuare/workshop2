#### Libraries used

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sys
```

### Read the csv file

```python
sys.path.append('../data/')

data = pd.read_csv('../data/spotify_dataset.csv')
data.head()
```

## Data Analysis

```python
# Popularity distribution
sns.histplot(data.popularity, kde=False, bins=30)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
# Convert duration from milliseconds to minutes
data.duration_min = data['duration_ms'] / 60000

# Plotting the histogram with a focus on tracks under 10 minutes
plt.figure(figsize=(10, 6))
sns.histplot(data[data.duration_min <= 15]['duration_min'], bins=30, kde=False)
plt.title('Distribution of Track Duration (Up to 10 minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

```python
# Select only the numeric columns
numeric_data = data.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
# Energy vs. Loudness
sns.scatterplot(x='energy', y='loudness', hue='explicit', data=data)
plt.title('Energy vs. Loudness by Explicitness')
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.show()
```

```python
# Explicit vs. Non-explicit content comparison
sns.countplot(x='explicit', data=data, palette='pastel', hue='explicit')
plt.title('Distribution of Explicit vs. Non-Explicit Tracks')
plt.xlabel('Explicit Content')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Explicit', 'Explicit'])
plt.show()
```

```python
## Distribution of mode
sns.countplot(x='mode', data=data, palette='muted', hue='mode')
plt.title('Distribution of Mode (Major/Minor)')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.show()
```
