## Data Analysis

```python
# Popularity distribution
sns.histplot(data.popularity, kde=False, bins=30)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
# Duration distribution
sns.histplot(data.duration_ms/60000, kde=False, bins=60)
plt.title('Distribution of Track Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between energy and loudness
sns.scatterplot(data=data, x='energy', y='loudness', hue='explicit')
plt.title('Energy vs. Loudness by Explicitness')
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
# Explicit vs. Non-explicit content comparison
sns.countplot(x='explicit', data=data, palette='pastel', hue='explicit')
plt.title('Distribution of Explicit vs. Non-Explicit Tracks')
plt.xlabel('Explicit Content')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Explicit', 'Explicit'])
plt.show()
```

```python
# Distribution of keys
sns.countplot(x='key', data=data, palette='muted', hue='key')
plt.title('Distribution of Musical Keys')
plt.xlabel('Key')
plt.ylabel('Count')
plt.show()
```

```python
## Distribution of mode
sns.countplot(x='mode', data=data, palette='muted', hue='mode')
plt.title('Distribution of Mode (Major/Minor)')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.show()
```

```python
# Top genres
top_genres = data.track_genre.value_counts().head(10)
sns.barplot(y=top_genres.index, x=top_genres.values, palette='viridis', hue=top_genres.values)
plt.title('Top 10 Genres')
plt.xlabel('Number of Tracks')
plt.ylabel('Genre')
plt.show()
```
