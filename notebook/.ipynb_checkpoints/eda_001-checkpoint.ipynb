# Exploratory data analisis of grammy awards dataset

#### Libraries used

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sqlalchemy import create_engine, text
import sys
import psycopg2
from datetime import datetime
```

#### Connection to the database

```python
 sys.path.append('../config/')
from config import configuration

config = configuration('../config/database.ini')
db_url = f"postgresql+psycopg2://{config['user']}:{config['password']}@{config['host']}/{config['database']}"

engine = create_engine(db_url)
```

#### Creating the table in the dataset

```python
create_table = text("""
CREATE TABLE IF NOT EXISTS the_grammy_award (
    id SERIAL PRIMARY KEY,
    year INT,
    title TEXT,
    published_at TIMESTAMP WITH TIME ZONE,
    updated_at TIMESTAMP WITH TIME ZONE,
    category TEXT,
    nominee TEXT,
    artist TEXT,
    workers TEXT,
    img TEXT,
    winner BOOLEAN
);
""")

with engine.begin() as connection:
    connection.execute(create_table)
```

#### Insert the information into the table

```python
data = pd.read_csv('../data/the_grammy_awards.csv')
data.to_sql('the_grammy_award', engine, if_exists='replace', index=False)
```

## We will just convert the types of columns that do not have the correct data type.

```python
data['published_at'] = pd.to_datetime(data['published_at'], utc=True).dt.tz_localize(None)
data['updated_at'] = pd.to_datetime(data['updated_at'], utc=True).dt.tz_localize(None)

print(data.dtypes)
```

```python
# Check duplicate entries

print(data.duplicated().sum())
```

## Data Analysis

```python
print(data.describe(include='all'))
```

```python
#Grammy Awards trend over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='year', estimator=len, data=data)
plt.title('Grammy Awards trend over time')
plt.xlabel('Year')
plt.ylabel('Number of awards')
plt.show()
```

```python
# Its the same graphic but this have year by year
plt.figure(figsize=(10, 6))
sns.countplot(x='year', data=data, palette='flare', hue='year')
plt.title('The Grammys distribution per years')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Number of awards')
plt.show() 
```

```python
# Top categories by Grammy nominations
plt.figure(figsize=(15, 10))
top_categories = data.category.value_counts().head(20)
sns.barplot(y=top_categories.index, x=top_categories.values, palette='viridis', hue=top_categories.values)
plt.title('Top 20 categories in the Grammys')
plt.xlabel('Number of Nominations')
plt.ylabel('Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Top 10 artists with most nominations
plt.figure(figsize=(12, 6))
top_artists = data.nominee.value_counts().head(10)
sns.barplot(x=top_artists.values, y=top_artists.index, palette='crest', hue=top_artists.values)
plt.title('Top 10 Artists with the Most Grammy Nominations')
plt.xlabel('Number of Nominations')
plt.ylabel('Artist')
plt.show()
```
